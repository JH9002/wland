```python
import pandas as pd
from teanaps.text_analysis import DocumentSummarizer

ds = DocumentSummarizer()
```

```python
PATH = '/content/drive/MyDrive/텍스트 마이닝_텀프로젝트/3차 데이터/코스피_정제_3차_20201101_20211031.csv'
```

```python
df_주식_1 = pd.read_csv(PATH, sep= ',')
content = df_주식_1['Content']
```

```python
len(content)
```

```python
SUMMARIZER_TYPE = "textrank"
MAX_SENTENCES = 3
```

```python
result_1 = []
count = 1

for i in content:
    count += 1
    print(count)
    result = ds.summarize(SUMMARIZER_TYPE, MAX_SENTENCES, document=i)
    result = ''.join(result)
    result=[result]
    result_1.append(result)
```

```python
df_1 = pd.DataFrame(result_1)
```
