# [형태소 분석]

```python
# 긍정

# TEANAPS 라이브러리를 불러옵니다.
from teanaps.nlp import MorphologicalAnalyzer ###형태소분석기
from teanaps.nlp import Processing ###전처리

processing = Processing() 
ma = MorphologicalAnalyzer()
#ma.set_tagger("okt")


tokenized_sentence_list_up = []
POS_LIST = ["NNG", "NNP"]

# PATH = "/content/drive/MyDrive/Colab_Notebooks/텍스트 마이닝/텍스트 마이닝 프로젝트★/sentence_token_Kospi9.txt"
# f = open(PATH, "w", encoding="utf-8")

# 파일에서 기사본문만 추출하여 명사형 문장을 생성합니다.
for line in a["Content"]: 
    line = line.strip()

    tagged_word_list = ma.parse(line) ### 형태소 분석 결과 저장
    tokenized_sentence = processing.get_plain_text(tagged_word_list, pos_list=POS_LIST, tag=False) 
    tokenized_sentence_list_up.append(tokenized_sentence)
# f.close()

tokenized_sentence_list_up[:3]
```

```python
edge_list = co.get_edge_list()[:20]
edge_list
```

```python
node_list = co.get_node_list()
node_list
```

```python
co_word_list = co.get_co_word("코스피") 
co_word_list
```

```python
centrality_dict = co.get_centrality("d_cent")
co.get_word_network_graph(centrality_dict, mode="markers+text", 
                          centrality_th=0.1, weight_th=10, node_size_rate=20, edge_width_rate=15,
                          ego_node_list=["코스피", "상승"])
```
