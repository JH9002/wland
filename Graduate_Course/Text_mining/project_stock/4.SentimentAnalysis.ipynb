# 감성분석

```python
import pandas as pd
```

```python
from teanaps.text_analysis import SentimentAnalysis

senti = SentimentAnalysis(model_path="/content/drive/MyDrive/텍스트마이닝 수업/WEEK 00/senti_model",
                          kobert_path="/content/drive/MyDrive/텍스트마이닝 수업/WEEK 00/mxnet_kobert")
```

```python
PATH = '/content/drive/MyDrive/텍스트 마이닝_텀프로젝트/5차 데이터/주식_요약.csv'
```

```python
df = pd.read_csv(PATH, sep=',', encoding='cp949')
```

```python
df
```

```python
content = df['Content']
# content = content
len(content)
```

```python
weight = []
count = 0
for sentence in content:
    count += 1
    print(count)

    result = senti.tag(sentence, neutral_th=0)
    if result[1] == 'positive':
        weight.append(result[0][1])
    else:
        weight.append(result[0][0])
```

```python
df['score'] = weight
```

```python
df
```

```python
save_path ='/content/drive/MyDrive/텍스트 마이닝_텀프로젝트/5차 데이터/주식_요약_감성.csv'
```

```python
df.to_csv(save_path)
```
