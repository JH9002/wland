# <예측을 위한 income(수익) 컬럼 데이터 생성>

```python
df['income'] = df['gross']- df['budget']
```

# <라벨 생성>

```python
income_label =[]

for i in list(df.income):
    if i>0:
        income_label.append(1)
    elif i == 0:
        income_label.append(-1)
    else:
        income_label.append(-1)
        
```

```python
df[ "income_label" ] = income_label
```

```python
income_results = []
for i in list(df.income_label):
    if i == 1:
        income_results.append("이익")
    else:
        income_results.append("손해")
```

```python
df["income_results"] = income_results
```

# <머신러닝 - 모델별 예측 및 평가>

```python
x_movies = df[['score','votes','budget']]
y_movies = df[['income_label']]
```

# Logistic regression(LR)

```python
from sklearn.linear_model import LogisticRegression
from sklearn.model_selection import train_test_split

x_train, x_test, y_train, y_test = train_test_split(x_movies, y_movies, random_state = 0)

clf1 = LogisticRegression().fit(x_train, y_train)
print('Accuracy of Logistic regression classifier on training set:{:.2f}'. format(clf1. score(x_train, y_train)))
print('Accuracy of Logistic regression classifier on test set:{:.2f}'. format(clf1. score(x_test, y_test),2))
```

# 서포트 벡터 머신(SVM)

```python
from sklearn.svm import SVC

x_train, x_test, y_train, y_test = (train_test_split(x_movies, y_movies, random_state = 0))

clf2 = SVC(C=10).fit(x_train, y_train)

print("Accuracy of Support Vector Machine classifier on training set:", str(round(clf2.score(x_train, y_train),2)))
print("Accuracy of Support Vector Machine classifier on test set:", str(round(clf2.score(x_test, y_test),2)))
```

# 디시전 트리(DT)

```python
from sklearn.tree import DecisionTreeClassifier

x_train, x_test, y_train, y_test = (train_test_split(x_movies, y_movies, random_state = 0))

clf3 = DecisionTreeClassifier().fit(x_train, y_train)

print("Accuracy of DecisionTree classifier on training set :", str(round(clf3.score(x_train, y_train),3)))
print("Accuracy of DecisionTree classifier on test set:", str(round(clf3.score(x_test, y_test),2)))
```

# 랜덤포레스트(RF)

```python
from sklearn.ensemble import RandomForestClassifier

x_train, x_test, y_train, y_test = (train_test_split(x_movies, y_movies, random_state = 0))

clf4 = RandomForestClassifier().fit(x_train, y_train)

print("Accuracy of RandomForest classifier on training set:", str(round(clf4.score(x_train, y_train),3)))
print("Accuracy of RandomForest classifier on test set:", str(round(clf4.score(x_test, y_test),2)))
```

# K-Nearest neighbor(KNN)

```python
from sklearn.neighbors import KNeighborsClassifier

clf5=KNeighborsClassifier().fit(x_train, y_train)

print("Accuracy of RandomForest classifier on training set:", str(round(clf5.score(x_train, y_train),3)))
print("Accuracy of RandomForest classifier on test set:", str(round(clf5.score(x_test, y_test),2)))
```

# <모델별 정확도 비교>

```python
LR_score = clf1.score(x_test,y_test)
SVM_score = clf2.score(x_test,y_test)
DT_score = clf3.score(x_test,y_test)
RF_score = clf4.score(x_test,y_test)
KNN_score = clf5.score(x_test,y_test)
```

```python
model = ['LR', 'SVM', 'DT', 'RF', 'KNN']
test_score = [LR_score, SVM_score, DT_score, RF_score, KNN_score]
plt. bar(model, test_score)
plt. show()
```
