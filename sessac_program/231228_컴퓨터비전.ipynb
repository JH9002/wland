```python
# file 경로를 |문자를, 특수제어기능으로 인지하기 때문에 
# r을 앞에 넣어서 디렉토리 구분자라고 명확하게 설명함

# from PIL import Image
file="C:/Users/jh/OneDrive - 연세대학교 (Yonsei University)/000 취업사관학교/새싹 프로그램) 강의 시작/img/dog.jpeg"

img = Image.open(file)
img.size #너비 408 높이 308 / 행은 308개 열은 408개
# [[1,3,4],[3,4,5]] 이런개 308개
print(type(img))
```

```python
# 이미지를 벡터화함
# 인공지능은 벡터자료만 취급하기때문
# PIL에서 제공되지 않는 작업중 내맘대로 이미지 변형하고자 할때
import numpy as np
imgArr = np.array(img)
# print(type(imgArr))
imgArr.shape

tmp=imgArr[:270,:,:] # 0~99번 행까지 100개의 행의 자료중, 모든 열, 모든 RGB 다 보여줘
tmp=imgArr[:,100:300,:]
imgArr[:,100:300,:]=0 # 검정으로 채우게됨, [255,0,0] 빨간색
tmp=imgArr
```
