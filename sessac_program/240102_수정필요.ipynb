```python
# 함수
def mysum(a):
    sum=0
    for i in a:
        sum+=i
    return sum

a=[1,2,3,4]
print(mysum(a))
```

```python
def myprn(fun):
    fun()

@myprn # prn함수를 myprn 함수안에서 실행하겠다
def prn():
    print('-'*50)

@myprn
def title():
    print('''
            ver=3.0
          ''')

```

```python
# 데코레이터 함수
def my_decorator(func):
    def wrapper():
        print("Something is happening before the function is called.")
        func()  # 원래 함수 호출
        print("Something is happening after the function is called.")
    return wrapper

# 데코레이터를 사용하여 함수 장식
@my_decorator
def say_hello():
    print("Hello!")

# 데코레이터가 추가된 함수 호출
say_hello()
```

```python
import qrcode

# QR 코드에 담길 데이터 # 사이트 1개
data = "https://www.naver.com"

# QR 코드 생성
qr = qrcode.QRCode(
    version=1,  # QR 코드의 버전 (1부터 40까지의 값)
    error_correction=qrcode.constants.ERROR_CORRECT_L,  # 오류 수정 레벨 (L, M, Q, H 중 선택)
    box_size=10,  # 각 박스의 크기
    border=4,  # 테두리의 크기
)
qr.add_data(data)
qr.make(fit=True)

# QR 코드 이미지 생성
img = qr.make_image(fill_color="black", back_color="white")

# 생성된 QR 코드 이미지 저장
img.save("c:/data/result/example_qrcode.jpg")
```

```python
import qrcode

# QR 코드에 담길 데이터 # 사이트 2개
data = ["https://www.naver.com", "https://www.daum.net"]  # 사이트 여러개일때
names = ['네이버', '다음']

# 각각의 데이터에 대해 QR 코드 생성
for site, name in zip(data, names):
    qr = qrcode.QRCode( # qrcode.make() 가능
        version=1,  # QR 코드의 버전 (1부터 40까지의 값)
        error_correction=qrcode.constants.ERROR_CORRECT_L,  # 오류 수정 레벨 (L, M, Q, H 중 선택)
        box_size=10,  # 각 박스의 크기
        border=4,  # 테두리의 크기
    )
    qr.add_data(site)
    qr.make(fit=True)

    # QR 코드 이미지 생성
    img = qr.make_image(fill_color="black", back_color="white")

    # 생성된 QR 코드 이미지 저장
    img.save("c:/data/result/"+ name+"_qrcode.jpg")
```
