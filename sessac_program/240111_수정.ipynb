```python
import numpy as np
import matplotlib.pyplot as plt

# 평균과 표준편차를 지정합니다.
mu = 0  # 평균
sigma = 1  # 표준편차

# 정규분포에 맞는 난수 1000개를 생성합니다.
random_numbers = np.random.normal(mu, sigma, 1000)

# 히스토그램을 그립니다.
plt.hist(random_numbers, bins=30, density=True, alpha=0.7)
plt.xlabel('값')
plt.ylabel('빈도')
plt.title('정규분포에 맞는 난수 히스토그램')
plt.grid(True)
plt.show()
```
