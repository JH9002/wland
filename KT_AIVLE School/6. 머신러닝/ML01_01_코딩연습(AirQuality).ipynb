# 3.데이터 준비

- **전처리** 과정을 통해 머신러닝 알고리즘에 사용할 수 있는 형태의 데이터를 준비합니다.

<img src = "https://github.com/Jangrae/img/blob/master/code_13.png?raw=true" width=700 align="left"/>

**1) 결측치 처리**

- 결측치가 있으면 제거하거나 적절한 값으로 채웁니다.

```python
# 결측치 확인
data.isna().sum()
```

```python
# 전날 값으로 결측치 채우기
data['Solar.R'].fillna(method='ffill', inplace=True) # 결측치 하나니까 slolar변수로 꼭안해도된다

# 확인
data.isna().sum()
```

**2) 변수 제거**

- 분석에 의미가 없다고 판단되는 변수는 제거합니다.

```python
# 변수 제거
drop_cols=['Month', 'Day']
data.drop(drop_cols, axis=1, inplace=True)

# 확인
data.head()
```

**3) x, y 분리**

- 우선 target 변수를 명확히 지정합니다.
- target을 제외한 나머지 변수들 데이터는 x로 선언합니다.
- target 변수 데이터는 y로 선언합니다. 
- 이 결과로 만들어진 x는 데이터프레임, y는 시리즈가 됩니다.
- 이후 모든 작업은 x, y를 대상으로 진행합니다.

```python
# target 확인
target = 'Ozone'

# 데이터 분리
x = data.drop(target, axis=1)  #★
y = data[target] # data.loc[:,target] #★

display(x,y)
```

**4) 학습용, 평가용 데이터 분리**

- 학습용, 평가용 데이터를 적절한 비율로 분리합니다.
- 반복 실행 시 동일한 결과를 얻기 위해 random_state 옵션을 지정합니다.

```python
# 모듈 불러오기
from sklearn.model_selection import train_test_split #★

# 7:3으로 분리
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=0.3, random_state=1) #★ stratify★
display(x_train.head(), x_test.head(), y_train.head(), y_test.head())

# x_train train데이터 셋의 x값 / 결국 x값이 젤중요
## 시계열 데이터는 섞으면 안되는 경우 있음 / 순서대로, shuffle=False 섞지마라, 인덱스 안섞임 #★
```

# 4.모델링

- 본격적으로 모델을 **선언**하고 **학습**하고 **평가**하는 과정을 진행합니다.
- 우선 **회귀** 문제인지 **분류** 문제인지 명확히 구분합니다.

<img src = "https://github.com/Jangrae/img/blob/master/code_14.png?raw=true" width=700 align="left"/>

**1) 모델링**

- 회귀 문제 인가요? 분류 문제인가요?
- 회귀인지 분류인지에 따라 사용할 알고리즘과 평가 방법이 달라집니다.
- 우선 다음 알고리즘과 평가 방법을 사용합니다.
    - 알고리즘: LinearRegression
    - 평가방법: mean_absolute_error

```python
# 1단계: 불러오기
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_absolute_error
```

```python
# 2단계: 선언하기
model = LinearRegression()
```

```python
# 3단계: 학습하기
model.fit(x_train, y_train)
```

```python
# 4단계: 예측하기
y_pred = model.predict(x_test)
```

```python
# 실제값, 예측값 비교
print(y_test.values[:10])
print(y_pred[:10])
```

```python
# 5단계: 평가하기
print('MAE:', mean_absolute_error(y_test, y_pred)) # 평균 13.9정도 차이있다
```

```python
# 기준 모델 성능 평가
mean_ozone = y_test.mean()
y_base = np.array([mean_ozone] * len(y_test)) # 배열로 해야되면 # [mean_ozone] * len(y_test) #이 형태 만들기★
```

```python
print('MAE:', mean_absolute_error(y_test, y_base)) 
```
