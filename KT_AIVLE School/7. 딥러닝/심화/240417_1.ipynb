```python
model = YOLO() # 기본 object detection / 코랩 왼쪽 폴더에 yolov8n.pt확인
```

```python
# 코랩 왼쪽 폴더의 data.yaml 들어가서 아래와 같이 변경
# train: ./train/images
# val: ./valid/images
```

```python
# gpu로 안하면 오래걸림
model.train(model='/content/yolov8n.pt',
            data='/content/Construction-Site-Safety-30/data.yaml',
            epochs=10,
            )


# nc 80, nc 25 확인
```

```python
model.predict(source='/content/Construction-Site-Safety-30/test/images/',
              conf=0.25,
              iou=0.7,
              save=True,
              )
```
