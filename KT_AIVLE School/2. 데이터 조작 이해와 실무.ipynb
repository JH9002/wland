### resample
- 일별<->월별 데이터 변환

```python
np.random.seed(1)
arr = np.random.randn(365)
time_idx = pd.date_range('2021-1-1', periods=365, freq='D') # 2021년 365일치 기간생성
ts = pd.Series(arr, index=time_idx)
ts
```

### down sampling

```python
ts.resample('M').mean() # 일별 데이터를 월별로 변환 # ★
```

### up sampling

```python
# 많이 사용안함, 1년치를 12달로 만들어야되는데 결측치를 사용자 정의따라 만들어야되므로 신뢰도 떨어질 우려있기 때문
arr = np.random.randn(10)
time_idx = pd.date_range('2021-1-1', periods=10, freq='W') # 주별 데이터 10개 생성
ts = pd.Series(arr, index=time_idx)
ts
```

```python
# forward filling
ts.resample('D').ffill().head(21) # 각 기간의 첫일을 참고하여 결측값 보간
```

```python
# backward filling
ts.resample('D').bfill().head(21) # 각 기간의 마지막날을 참고하여 결측값 보간
```

## icecream 데이터셋 로딩

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
# 데이터 생성
np.random.seed(1) # 난수 설정
arr = np.random.randint(8000, 20000, 167) # 8000~20000 167개의 데이터 생성

# datafrmae columns
_columns = ["year", "month", "date", "series_amount"]

# 날짜 랜덤 생성
date = pd.date_range(start="2008-01-01", end="2021-12-31", periods=167)
MONTH = ["Jan", "Feb", "Mar", "Apr", "May", "Jun", "Jul", "Aug", "Sep", "Oct", "Nov", "Dec"]
ice_df = pd.DataFrame(columns=_columns)

for idx in range(len(date)):
    year = date[idx].year
    month = MONTH[date[idx].month -1]
    day = date[idx].day
    amount = arr[idx]
    
    array = [[year, month, day, amount]]
    df = pd.DataFrame(array, columns=_columns)
    ice_df = pd.concat([ice_df, df])
ice_df.reset_index(inplace=True)
ice_df.drop("index", axis=1, inplace=True)
print(ice_df)
```

```python
# ice_df = pd.read_csv('./icecream.csv')
# ice_df.info()
# ice_df.index

ice_df['year']=ice_df['year'].astype(str)
ice_df['month']=ice_df['month'].astype(str)
ice_df['date']=ice_df['date'].astype(str)
# ice_df.info()

# 날짜 문자열 생성 : 2020-01-01 형태
date_str = ice_df['year'].values + '-' + ice_df['month'].values + '-'+ ice_df['date'].values
# ice_df[:5]

ice_idx = pd.to_datetime(date_str)
# ice_idx

ice_df.index = ice_idx
# ice_df

ice_df.drop(['year', 'month', 'date'], axis=1, inplace=True)
# ice_df

ice_df.plot(figsize=(8,3))
plt.show()
```
