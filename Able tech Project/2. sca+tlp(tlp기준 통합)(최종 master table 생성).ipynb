```python
'''
<for문>
-둘다 있으면 아무것도 안한다
-sca값 널이아니면 그값을 tlp에다가 넣는다 
-sca널이면 아무것도 안한다
'''
```

```python
# 221123 수정
sca_list = ['self_efficacy', 'math_self_efficacy', 'in_goal_Orientation','out_goal_Orientation', 'test_anxiety', 'parent_communication', 'academy_relies']
tlp_list = ['Self_Efficacy', 'Math_efficacy', '숙달접근', '수행접근', '총불안', '전체관계', '학원의존' ]

for i in range(1902):
    for j in range(7):
        if pd.isna(df.loc[i, sca_list[j]]) != True:            ## sca 값 있을때
            if pd.isna(df.loc[i, tlp_list[j]]) != True:        ## tlp 값도 있으면
                continue    ## 넘어감


            else:           ## sca있고 tlp값은 없으면
                df.loc[i, tlp_list[j]] = df.loc[i, sca_list[j]] ## tlp에 sca 값 넣기  


        else:              ## sca 값 없으면
            continue      ## 넘어감


df[['Self_Efficacy', 'Math_efficacy', '숙달접근', '수행접근', '총불안', '전체관계', '학원의존' ]]
```

```python
# tlp_fear : 시험불안 + 부모자녀의사소통 + 학원의존
#          * 시험불안 = (정서불안, 인지불안, 관계불안, 일반불안, 총불안)

# 총불안 :  시험불안 4개항목(정서불안, 인지불안, 관계불안, 일반불안) 평균값
# sca에서 tlp로 합친 데이터들은 시험불안 항목 5개 값이 다 있지 않고, 총불안(평균값) 값만 있기 때문에 총불안x4값으로 계산후 평균계산

df[['총불안']]= df[['총불안']]*4
df[['총불안']]
```

```python
tlp_fear = df[['총불안','부모관계긍정', '부모관계부정', '학원의존']]
tlp_fear
```

```python
tlp_fear.describe()
```

```python
from pandas import Series
h='부모관계긍정'

for i in range(1902):
    if pd.isna(tlp_fear.loc[i, h]) != True: # 부모관계긍정 값이 있으면
        tlp_fear.loc[i, 'tlp_fear'] = (tlp_fear.sum(axis=1)[i]) / 7
    else:
        tlp_fear.loc[i ,'tlp_fear'] = (tlp_fear.sum(axis=1)[i]) / 5
tlp_fear
```

```python
tlp_fear.max()
```

```python
df['tlp_fear'] = tlp_fear['tlp_fear']
df['tlp_fear']
```
