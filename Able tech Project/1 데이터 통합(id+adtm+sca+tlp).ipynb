```python
'''
<for문>
-둘다 있으면 아무것도 안한다
-sca값 널이아니면 그값을 tlp에다가 넣는다 
-sca널이면 아무것도 안한다
'''
```

```python
sca_list = ['self_efficacy', 'math_self_efficacy', 'in_goal_Orientation','out_goal_Orientation', 'test_anxiety', 'parent_communication', 'academy_relies']
tlp_list = ['Self_Efficacy', 'Math_efficacy', '숙달접근', '수행접근', '총불안', '전체관계', '학원의존' ]

for i in range(1902):
    for j in range(7):
        if pd.isna(sca_scaled.loc[i, sca_list[j]]) != True:     ## sca 값 있을때
            if pd.isna(tlp_scaled.loc[i, tlp_list[j]]) != True: ## tlp 값도 있으면
                continue    ## 넘어감


            else:           ## sca있고 tlp값은 없으면
                tlp_scaled.loc[i, tlp_list[j]] = sca_scaled.loc[i, sca_list[j]] ## tlp에 sca 값 넣기  


        else:              ## sca 값 없으면
            continue      ## 넘어감


tlp_scaled[['Self_Efficacy', 'Math_efficacy', '숙달접근', '수행접근', '총불안', '전체관계', '학원의존' ]]
```

```python
df[['Self_Efficacy', 'Math_efficacy', '숙달접근', '수행접근', '총불안', '전체관계', '학원의존' ]] = tlp_scaled[['Self_Efficacy', 'Math_efficacy', '숙달접근', '수행접근', '총불안', '전체관계', '학원의존' ]]
df
```
